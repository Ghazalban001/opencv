```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create a 300x300 black canvas
square=np.zeros((300,300),np.uint8)
# Draw a filled white rectangle on the canvas
cv2.rectangle(square, (50, 50), (250, 250), 255, -1)

# Create another 300x300 black canvas
ellipse=np.zeros((300,300),np.uint8)
# Draw a filled ellipse on the canvas
cv2.ellipse(ellipse,(150,150),(150,150),30,0,180,225,-1)


# Display the square image
plt.subplot(121);plt.imshow(square,cmap="gray");plt.title("square")
# Display the ellipse image
plt.subplot(122);plt.imshow(ellipse,cmap="gray");plt.title("ellipse")
```

```python
# Perform bitwise AND between the two shapes
bitwise_and=cv2.bitwise_and(square,ellipse)
# Perform bitwise OR between the two shapes
bitwise_or=cv2.bitwise_or(square,ellipse)
# Perform bitwise XOR between the two shapes
bitwise_xor=cv2.bitwise_xor(square,ellipse)
# Invert the square image
bitwise_not_sq=cv2.bitwise_not(square)
```

```python
# create a figure for showing all results
plt.figure(figsize=[12,6])
# display original square image
plt.subplot(242);plt.imshow(square,cmap="gray");plt.title("square")
# display original ellipse image
plt.subplot(243);plt.imshow(ellipse,cmap="gray");plt.title("ellipes")
# display bitwise AND result
plt.subplot(245);plt.imshow(bitwise_and,cmap="gray");plt.title("AND")
# display bitwise OR result
plt.subplot(246);plt.imshow(bitwise_or,cmap="gray");plt.title("OR")
# display bitwise XOR result
plt.subplot(247);plt.imshow(bitwise_xor,cmap="gray");plt.title("XOR")
# display inverted square image
plt.subplot(248);plt.imshow(bitwise_not_sq,cmap="gray");plt.title("not")
```

```python
# Load the original image
image = cv2.imread("tabiat.jpg")

# Crop a region from the image
cropped = image[100:600, 150:650]
plt.imshow(cropped[...,::-1])  # display cropped image in RGB format

# Create a black image with the same size as the cropped image
circle = np.zeros((500, 500, 3), "uint8")

# Draw a filled white circle on the black image
cv2.circle(circle, (250, 250), 250, (255, 255, 255), -1)

# Apply the circular mask to the cropped image
output_image = cv2.bitwise_and(cropped, circle)

# Display the masked output image
plt.imshow(output_image[...,::-1])
```
