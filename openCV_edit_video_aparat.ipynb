```python
import cv2
import yt_dlp

url = "https://www.aparat.com/v/dyq8698"

# تنظیمات دانلود
ydl_opts = {
    #«بهترین کیفیت ممکن را دانلود کن.
    #اگر MP4 وجود داشت، MP4 را ترجیح بده.
    #بهترین فایل که پسوندش mp4 باشد.
    "format": "best[ext=mp4]/best",  
    "outtmpl": "video.%(ext)s"   
}

#یک شیء دانلودر می‌سازد. به اسم ydl
with yt_dlp.YoutubeDL(ydl_opts) as ydl:
    ydl.download([url])

# باز کردن ویدیو
cap = cv2.VideoCapture("video.mp4")

if not cap.isOpened():
    print("خطا: ویدیو باز نشد")
    exit()

# مشخصات ویدیو
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

#سرعت ویدیو
fps = cap.get(cv2.CAP_PROP_FPS)
if fps == 0:
    fps = 30   # مقدار پیش‌فرض امن

print("width:", width)
print("height:", height)
print("fps:", fps)

# ذخیره خروجی
fourcc = cv2.VideoWriter_fourcc(*"mp4v")
out = cv2.VideoWriter("output.mp4", fourcc, fps, (width, height))

while True:
    ret, frame = cap.read()

    if not ret:
        break

    # تبدیل به سیاه سفید
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # تبدیل دوباره به BGR برای ذخیره
    result = cv2.cvtColor(gray, cv2.COLOR_GRAY2BGR)

    out.write(result)

    cv2.imshow("Video", result)

    if cv2.waitKey(1) & 0xFF == ord("q"):
        break

cap.release()
out.release()
cv2.destroyAllWindows()

print("تمام شد")
```
