```python
import cv2
import numpy as np

# create a black canvas
img = np.zeros((512, 512, 3), np.uint8)  
# state to track mouse button status
drawing = False  
# coordinates where the drawing starts
start_point = (0, 0)

def drawing_rectangle(event, x, y, flags, params):
    global start_point, drawing  
    if event == cv2.EVENT_LBUTTONDOWN:
        drawing = True  # start drawing
        start_point = (x, y)   # record start position
    elif event == cv2.EVENT_MOUSEMOVE:
        if drawing == True:
            # draw rectangle as mouse moves
            cv2.rectangle(img, start_point, (x, y), (0, 255, 0), -1)
    elif event == cv2.EVENT_LBUTTONUP:
        drawing = False  # stop drawing

cv2.namedWindow("image")
cv2.setMouseCallback("image", drawing_rectangle)  # bind mouse event to window

while True:
    cv2.imshow("image", img)
    k = cv2.waitKey(1) & 0xFF
    if k == 27:  # press ESC to exit
        break
cv2.destroyAllWindows()
```
