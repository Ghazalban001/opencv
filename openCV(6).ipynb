```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img=plt.imread("tabiat.jpg")
b,g,r=cv2.split(img)
```

- u یعنی Unsigned (بدون علامت مثبت و منفی، فقط مثبت).
- int یعنی Integer (عدد صحیح، نه اعشاری).
- 8 یعنی ۸ بیتی (یعنی اعداد فقط می‌توانند بین ۰ تا ۲۵۵ باشند).
- بیت یعنی 0و1 بودن 
- هر 8 بیت ---> 1 بایت
- هر 1 بایت می تواند یک اطلاعات رو ذخیره کنه



```python
#create a zero matris
#یک ماتریس صفر در صفر به اندازه کانال red
#ماتریس صفر یعنی عکس سیاه

# Create a zero matrix with same size as a channel
zeros=np.zeros_like(r,"uint8")

# Keep only red channel, others set to zero
red=cv2.merge([zeros,zeros,r])
# Keep only green channel
green=cv2.merge([zeros,g,zeros])
# Keep only blue channel
blue=cv2.merge([b,zeros,zeros])

# Show red , blue , green image
plt.figure(figsize=[20,5])
plt.subplot(141);plt.imshow(red[...,::-1],cmap='gray');plt.title("red");
plt.subplot(142);plt.imshow(green[...,::-1],cmap="gray");plt.title("green");
plt.subplot(143);plt.imshow(blue[...,::-1],cmap="gray");plt.title("blue");
```

```python
img=cv2.imread("tabiat.jpg")
b,g,r=cv2.split(img)
#زیاد کردن مقدارblue
modified=b+20
merged=cv2.merge([modified,g,r])

plt.subplot(141);plt.imshow(r,cmap="gray");plt.title("red");
plt.subplot(142);plt.imshow(g,cmap="gray");plt.title("green")
plt.subplot(143);plt.imshow(modified,cmap="gray");plt.title("blue")
plt.subplot(144);plt.imshow(merged[:,:,::-1]);plt.title("modifide blue channel")
```

```python
#حل مشکل رنگ های تصویر بعد از اضافه کردن آبی به شبکه
modified_b=b.astype("int16")+20
modified_b=np.minimum(modified_b,255)
modified_b=modified_b.astype('uint8')

merged=cv2.merge([modified_b,g,r])
plt.figure(figsize=[20,8])
plt.subplot(241);plt.imshow(r,cmap="gray");plt.title("Red Chnnel");
plt.subplot(242);plt.imshow(r,cmap="gray");plt.title("Red Chnnel");
plt.subplot(243);plt.imshow(g,cmap="gray");plt.title("orginal imge");
plt.subplot(244);plt.imshow(img[:,:,::-1]);plt.title("original image")

plt.subplot(247);plt.imshow(modified_b,cmap="gray");plt.title("Modified blue channel");
plt.subplot(248);plt.imshow(merged[:,:,::-1]);plt.title("modified Image")
```
