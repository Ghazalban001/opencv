```python
# Load an image using 'imread' specifying the path to image
img=cv2.imread("picture.jpeg")
```

```python
#تغییر و دستکاری پیکسل ها
import cv2
img=cv2.imread("picture.jpeg")
img[200:300,300:500,:]=255

#این قسمت کانال‌های رنگ را برعکس می‌کند
#... ---> انتخاب همه سطر و ستون ها 
plt.imshow(img[...,::-1])
```
