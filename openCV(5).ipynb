```python
img=cv2.imread("tabiat.jpg")

# Create large figure for display
plt.figure(figsize=[20,20])
# Split image into B, G, R channels
b,g,r=cv2.split(img)

# Show red channel in grayscale
plt.subplot(141);plt.imshow(r,cmap="gray");plt.title("red channel")
# Show green channel in grayscale
plt.subplot(142);plt.imshow(g,cmap="gray");plt.title("green channel")
# Show blue channel in grayscale
plt.subplot(143);plt.imshow(b,cmap="gray");plt.title("blue title")
# Show original image (convert BGR to RGB)
plt.subplot(144);plt.imshow(img[:,:,::-1]);plt.title("all channel")
```

merge channels from an image

```python
merge=cv2.merge([b,g,r])
plt.figure(figsize=[20,5])

# Show red channel
plt.subplot(141);plt.imshow(r,cmap="gray");plt.title("red channel")
# Show green channel
plt.subplot(142);plt.imshow(g,cmap="gray");plt.title("green channel")
# Show green channel
plt.subplot(143);plt.imshow(b,cmap="gray");plt.title("blue title")
# Show merged/original image (BGR → RGB for display)
plt.subplot(144);plt.imshow(img[:,:,::-1]);plt.title("all channel")
```
