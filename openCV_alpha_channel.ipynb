```python
import cv2
import matplotlib.pyplot as plt
```

```python
# Load the image keeping its original format (including alpha channel)
img=cv2.imread("minium.png",cv2.IMREAD_UNCHANGED)
# Get the image dimensions (height, width, channels)
img.shape
```

```python
plt.imshow(img)
```

```python
# Extract the first 3 channels (BGR)
brg_image=img[...,0:3]
# Extract the last channel (Alpha/Transparency)
alpha_image=img[...,-1]
# Convert the BGR format to RGB for correct display in matplotli
rgb_image=cv2.cvtColor(brg_image,cv2.COLOR_BGR2RGB)
# Display the resulting RGB imag
plt.imshow(rgb_image);plt.title("RGB")
```

```python
# Display the alpha channel in grayscale
plt.imshow(alpha_image,cmap="gray");plt.title("Alpha")
```

```python
# Read the image including the alpha channel
img=cv2.imread("minium.png",cv2.IMREAD_UNCHANGED)
img.shape
```

```python
# Separate the BGR color channels (first three channels)
bgr_image=img[...,0:3]
# Separate the Alpha (transparency) channel (last channel)
alpha_image=img[...,-1]
# Convert BGR format to RGB for correct color display
reg_image=cv2.cvtColor(bgr_image,cv2.COLOR_BGR2RGB)
# Show the final RGB image with a title
plt.imshow(reg_image);plt.title("RGB")
```
