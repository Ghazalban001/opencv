```python
import cv2

alpha_slider_max = 100
title_window = 'Linear Blend'

def on_trackbar(val):
    alpha = val / alpha_slider_max
    beta = (1.0 - alpha)
    # ترکیب دو عکس
    dst = cv2.addWeighted(src1, alpha, src2, beta, 0.0)
    cv2.imshow(title_window, dst)

# ۱. خواندن عکس‌ها
src1 = cv2.imread("tabiat.jpg")
src2 = cv2.imread("minion.jpg")

# ۲. چک کردن اینکه عکس‌ها وجود داشته باشند
if src1 is None or src2 is None:
    print("خطا: عکس‌ها پیدا نشدند! مطمئن شو نام فایل‌ها درست است.")
    exit()

# ۳. هم‌اندازه کردن عکس دوم با عکس اول (بسیار مهم)
src2 = cv2.resize(src2, (src1.shape[1], src1.shape[0]))

size = (500, 400)
src1 = cv2.resize(src1, size)
src2 = cv2.resize(src2, size)

# ۴. ساخت پنجره و نوار لغزنده
cv2.namedWindow(title_window)
trackbar_name = "Alpha"
cv2.createTrackbar(trackbar_name, title_window, 0, alpha_slider_max, on_trackbar)

# ۵. نمایش اولیه
on_trackbar(0)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import numpy as np
import cv2

def nothing(x):
    pass

# ایجاد یک تصویر مشکی
img = np.zeros((300, 512, 3), np.uint8)
cv2.namedWindow("image")

# ایجاد Trackbarها
cv2.createTrackbar("R", "image", 0, 255, nothing)
cv2.createTrackbar("G", "image", 0, 255, nothing)
cv2.createTrackbar("B", "image", 0, 255, nothing)

switch = "0 : OFF \n1 : ON"
cv2.createTrackbar(switch, "image", 0, 1, nothing)

while True:
    cv2.imshow("image", img)
    
    # خروج با زدن کلید ESC
    if cv2.waitKey(1) & 0xFF == 27:
        break

    # گرفتن موقعیت لحظه‌ای اسلایدرها
    r = cv2.getTrackbarPos("R", "image")
    g = cv2.getTrackbarPos("G", "image")
    b = cv2.getTrackbarPos("B", "image")
    s = cv2.getTrackbarPos(switch, "image")

    if s == 0:
        img[:] = 0
    else:
        img[:] = [b, g, r] # مقداردهی صحیح انجام شد

cv2.destroyAllWindows()
```
