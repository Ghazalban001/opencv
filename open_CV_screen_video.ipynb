```python
import numpy as np
from PIL import ImageGrab
import cv2
```

```python
while True:
    # گرفتن اسکرین‌شات از صفحه نمایش
    img = ImageGrab.grab()
    
    # تبدیل تصویر به آرایه عددی (Numpy Array)
    frame = np.array(img)

    # تبدیل رنگ از RGB به BGR (چون OpenCV تصویر را به صورت BGR نمایش می‌دهد)
    frame_corrected = cv2.cvtColor(frame, cv2.COLOR_RGB2BGR)

    # نمایش تصویر در پنجره
    cv2.imshow("Screen Capture", frame_corrected)

    # اگر کلید Esc (کد 27) فشار داده شد، حلقه را بشکن
    if cv2.waitKey(1) & 0xFF == 27:
        break

# بستن تمام پنجره‌ها
cv2.destroyAllWindows()
```
