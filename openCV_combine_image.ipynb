ترکیب تصویر های رنگی

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
image1=cv2.imread("girl.png") 
shape=image1.shape # get the shape of the girl image

# crop the jungle image with the same size as the girl image
image2=cv2.imread("jangle .jpg")[100:100+shape[0],100:100+shape[1],:]

output_image=cv2.add(image1,image2)

plt.figure(figsize=[10,5])   # create a figure with custom size
plt.subplot(131);plt.imshow(image1[...,::-1]);plt.title("image1")  # display first image
plt.subplot(132);plt.imshow(image2[...,::-1]);plt.title("image2")  # display second image
plt.subplot(133);plt.imshow(output_image[...,::-1]);plt.title("image1+image2") # display added result

```
