خواندن تصویر

```python
import cv2

cap=cv2.VideoCapture("atash.mp4")

while True:
    ret,frame=cap.read()

    if not ret:
        print("Exiting...")
        break
    cv2.imshow("video",frame)
    if cv2.waitKey(1)& 0xFF==27:
        break
cap.release()
cv2.destroyAllWindows()
```

سیاه و سفید کردن ویدیو

```python
import cv2

cap=cv2.VideoCapture("atash.mp4")

while True:
    ret,frame=cap.read()

    if not ret:
        print("Exiting...")
        break
    gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    cv2.imshow("Gray video",gray)
    if cv2.waitKey(1)& 0xFF==27:
        break
cap.release()
cv2.destroyAllWindows()
```
