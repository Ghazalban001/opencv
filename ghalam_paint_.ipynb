```python
import cv2
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
img=np.zeros((512,512,3),np.uint8)
drowing=False
ix=0
iy=0
```

```python
def draw(event, x, y, flags, params):
    global ix, iy, drowing
    
    # Start drawing when the left mouse button is pressed
    if event == cv2.EVENT_LBUTTONDOWN:
        drowing = True
        ix, iy = x, y

    # Draw a line while the mouse moves and the button is held down
    if event == cv2.EVENT_MOUSEMOVE:
        if drowing == True:
            # Draw a line from the previous point (ix, iy) to the current (x, y)
            # Note: 'thickness' was misspelled as 'thinckness' in your original code
            cv2.line(img, pt1=(ix, iy), pt2=(x, y), color=(255, 255, 255), thickness=3)
            ix, iy = x, y  # Update the starting point for the next segment

    # Stop drawing when the left mouse button is released
    if event == cv2.EVENT_LBUTTONUP:
        drowing = False

# 2. Setup: Create a named window and link it to the mouse callback function
cv2.namedWindow("Window")
cv2.setMouseCallback("Window", draw)

# 3. Execution Loop: Continuously update and show the image
while True:
    cv2.imshow("Window", img) # Ensure the window name matches namedWindow
    
    # Exit the loop if the 'ESC' key (ASCII 27) is pressed
    if cv2.waitKey(1) & 0xFF == 27:
        break

# 4. Cleanup: Close all GUI windows
cv2.destroyAllWindows()
```
